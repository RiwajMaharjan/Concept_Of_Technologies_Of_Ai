**Task 1**

```python
low = []
moderate = []
high = []
# Daily time (in hours): [study, entertainment, sleep]
time_data = [
(3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0),
(4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5),
(5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0),
(3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0),
(4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0),
]


def studyTimeClassifier(data):
    """
        Classifies study hours and appends them to global list: low, moderate and high.

        Parameters:
                list of tuples: Each tuple contains (study_hours, entertainment_hours, sleep_hours)
    """
    for i in data: #i is a tuple from the list

        if i[0] > 5: # if study hour is greater than 5.
            high.append(i[0])

        elif 3 <= i[0] <= 5: # if study hour is between 3 and 5.
            moderate.append(i[0])

        else: # if study hour is less than 3.
            low.append(i[0])

studyTimeClassifier(time_data) # calling the classifier function
print("Low study hours list: ", low)
print("Moderate study hours list: ", moderate)
print("High study hours list: ", high)

```

**Task 2**

```python
#Task 2
print("Number of low study hours list: ", len(low))
print("Number of moderate study hours list: ", len(moderate))
print("Number of high study hours list: ", len(high))
```

**Task 3**

```python
def convertStudyToMinutes(data):
    """
    Converts study hours to minutes and returns a new list.

    Parameters:
        data (list of tuples): Each tuple contains (study_hours, entertainment_hours, sleep_hours)

    Returns:
        list: Study hours converted to minutes
    """
    return [day[0] * 60 for day in data]


study_minutes = convertStudyToMinutes(time_data)
print("Study hours list in minutes: ", study_minutes)
```

**Task 4**

```python
study_hours = []
entertainment_hours = []
sleep_hours = []

def extractTimeUse(data):
    """
        Extracts study, entertainment, and sleep hours from the dataset
        and appends them to the corresponding global lists.

        Parameters:
                list of tuples: Each tuple contains (study_hours, entertainment_hours, sleep_hours)
    """
    for i in data:
        study_hours.append(i[0])
        entertainment_hours.append(i[1])
        sleep_hours.append(i[2])

def sumEverything(data):
  """
    Calculates the sum of a list of numbers.

    Parameters:
        data (list of floats/ints): List of numeric values to sum.

    Returns:
        float: Total sum of all numbers in the list.
  """
  total = 0
  for i in data:
    total += i
  return total

#Task 4
extractTimeUse(time_data)
avg_study = round(sumEverything(study_hours) / len(study_hours), 2)
avg_ent = round(sumEverything(entertainment_hours) / len(entertainment_hours), 2)
avg_sleep = round(sumEverything(sleep_hours) / len(sleep_hours), 2)

print("Average hours spent studying:", avg_study)
print("Average hours spent on entertainment:", avg_ent)
print("Average hours spent sleeping:", avg_sleep)
```
