**Problem - 4: Matrix Operations:**


```python
A = np.array([[3, 4],
              [7, 8]])
B = np.array([[5, 3],
              [2, 1]])


A_inv = np.linalg.inv(A)
I = np.dot(A, A_inv)
print("A · A⁻¹:", I)
```

```python
AB = np.dot(A, B)
BA = np.dot(B, A)

print("A·B: ", AB)
print()
print("B·A: ", BA)
print()
print("Are they equal?", np.array_equal(AB, BA))
```

```python
AB_T = np.dot(A, B).T
B_TA_T = np.dot(B.T, A.T)

print("(AB)ᵀ =\n", AB_T)
print()
print("Bᵀ·Aᵀ =\n", B_TA_T)
print("Are they equal?", np.array_equal(AB_T, B_TA_T))
```

```python
# 2x − 3y + z = −1
# x − y + 2z = −3
# 3x + y − z = 9
A_eq = np.array([[2, -3, 1],
                 [1, -1, 2],
                 [3, 1, -1]])

B_eq = np.array([-1, -3, 9])

X = np.dot(np.linalg.inv(A_eq), B_eq)
print("Solution [x, y, z]: ", X)

```

```python
X_solve = np.linalg.solve(A_eq, B_eq)
print("Solution using np.linalg.solve:", X_solve)
```

**Experiment: How Fast is Numpy?**

```python
import numpy as np
import time


# 1. Element-wise Addition

size = 1_000_000

# Python Lists
list1 = list(range(size))
list2 = list(range(size))

start = time.time()
list_add = [list1[i] + list2[i] for i in range(size)]  # element-wise addition
end = time.time()
print("Python list addition time:", end - start, "seconds")

# NumPy Arrays
arr1 = np.arange(size)
arr2 = np.arange(size)

start = time.time()
np_add = arr1 + arr2  # vectorized addition
end = time.time()
print("NumPy array addition time:", end - start, "seconds")


# 2. Element-wise Multiplication


# Python Lists
start = time.time()
list_mul = [list1[i] * list2[i] for i in range(size)]  # element-wise multiplication
end = time.time()
print("Python list multiplication time:", end - start, "seconds")

# NumPy Arrays
start = time.time()
np_mul = arr1 * arr2  # vectorized multiplication
end = time.time()
print("NumPy array multiplication time:", end - start, "seconds")

# 3. Dot Product

# Python Lists
start = time.time()
dot_list = sum(list1[i] * list2[i] for i in range(size))  # compute dot product
end = time.time()
print("Python list dot product time:", end - start, "seconds")

# NumPy Arrays
start = time.time()
dot_np = np.dot(arr1, arr2)  # fast dot product
end = time.time()
print("NumPy array dot product time:", end - start, "seconds")


# 4. Matrix Multiplication (1000x1000)

matrix_size = 100

# Python Lists (nested loops)
matrix1 = [[i + j for j in range(matrix_size)] for i in range(matrix_size)]  # create matrix1
matrix2 = [[i - j for j in range(matrix_size)] for i in range(matrix_size)]  # create matrix2

start = time.time()
# naive Python matrix multiplication (slow)
result_matrix = [[sum(matrix1[i][k] * matrix2[k][j] for k in range(matrix_size))
                  for j in range(matrix_size)]
                 for i in range(matrix_size)]
end = time.time()
print("Python list matrix multiplication time:", end - start, "seconds")

# NumPy Arrays
np_matrix1 = np.array(matrix1)
np_matrix2 = np.array(matrix2)

start = time.time()
np_result_matrix = np.dot(np_matrix1, np_matrix2)  # fast NumPy matrix multiplication
end = time.time()
print("NumPy matrix multiplication time:", end - start, "seconds")
```
